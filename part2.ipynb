```python
import pandas as pd

df = pd.read_csv('amazon-orders.csv')
df.head()
```

```python
df.shape
```

```python
df = df.fillna(0)
df.head()
```

```python
#take out $ for total charged column
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
#total amount of money spent
df["Total Charged"].sum()
```

```python
#average money spent per order
df["Total Charged"].mean()
```

```python
#medium money spent
df["Total Charged"].median()
```

```python
#highest priced order
df["Total Charged"].max()
```

```python
#lowerst priced order
df["Total Charged"].min()
```

```python
#take out $ for taxed charged column
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
#total money spent toward taxes
df["Tax Charged"].sum()
```

```python
#overall tax rate
df["Tax Charged"].sum() / df["Total Charged"].sum()
```

```python
#convert to datatime data type so the computer can recognize them as dates
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
df.plot.bar(x='Order Date', y='Total Charged', rot=90, figsize=(20,10))
```

```python
#grouping so that each bar represents each day not each order
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
#Total spending bar graph
daily_orders.plot.bar(figsize=(20,10), color='#61D199', xlabel='Order Date', ylabel='Spending', title='Amazon Spending Over Time')
```

```python
#grouping by year
yearly_spending = df.groupby(pd.Grouper(key='Order Date', freq='1Y')).sum()['Total Charged']
yearly_spending.head()
```

```python
#pie chart spending per year
import matplotlib.pyplot as plt

mylabels= ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

yearly_spending.plot.pie(y='Total Charged', labels = mylabels, title = 'Yearly Spending')
plt.legend(bbox_to_anchor=(1.5, 1.0), loc = 'upper right')
    
```

```python
#bar graph yearly spending
mylabels = ['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022'] 

yearly_spending.plot.bar(figsize=(10,5), color='g', xlabel='Year', ylabel='Total Spending', title='Yearly Amazon Spending')
```

```python
#multiple lines total charged vs tax charged

import matplotlib.pyplot as plt

plt.plot(df["Order Date"], df["Total Charged"])
plt.plot(df["Order Date"], df["Tax Charged"])

plt.title("Total Charged vs. Tax Charged")
plt.xlabel("Date")
plt.ylabel("Spending")
```
